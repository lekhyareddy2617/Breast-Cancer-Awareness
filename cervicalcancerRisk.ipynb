```python
'''
To build and train an XG-Boost classifier to predict whether a person has a risk of having cervical cancer. Cervical cancer kills about 4,000 women in the U.S. and about 300,000 women worldwide. 
Data has been obtained from 858 patients and include features such as number of pregnancies, smoking habits, Sexually Transmitted Disease (STD), demographics, and historic medical records.
'''
```

```python
#dataset: https://www.kaggle.com/loveall/cervical-cancer-risk-classification
import pandas as pd
import numpy as np

#data visulization libraries
import seaborn as sns
import matplotlib.pyplot as plt

import zipfile
import plotly.express as px

#to make sure axes are seen
from jupyterthemes import jtplot
jtplot.style(theme='monokai',context='notebook',ticks=True,grid=False)
```

```python
cancer_df=pd.read_csv('C:/Users/lekhy/OneDrive/Desktop/lekhya/cervical_cancer.csv')
cancer_df
```

```python
#cancer_df.info()
#cancer_df.describe()
cancer_df=cancer_df.replace('?',np.nan) #missing values marked w '?' replaced w NaN

#making heatmap to see places where null is present
plt.figure(figsize=(20,20))
sns.heatmap(cancer_df.isnull())
```

```python
#dropping the colums with many null values (STDs: Time since first diagnosis and STDs: Time since last diagnosis)
cancer_df=cancer_df.drop(columns=['STDs: Time since first diagnosis','STDs: Time since last diagnosis'])
cancer_df.info()
```

```python
#coverting 'object' to numeric type tht lets us apply operations
cancer_df=cancer_df.apply(pd.to_numeric) 
cancer_df.info()
```

```python
cancer_df.describe()
```

```python
#replacing NaN values with average of the row
cancer_df=cancer_df.fillna(cancer_df.mean())
cancer_df
```

```python
sns.heatmap(cancer_df.isnull())
```

```python
#correlation matrix
corr_matrix=cancer_df.corr()
corr_matrix #0 is no correlation; 1 is full correlation; negative value means inverse correlation
```

```python
#heatmap for correlation matrix
plt.figure(figsize=(30,30))
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

```python
cancer_df.hist(bins=10, figsize= (30,30), color='yellow')
```

```python
target_df=cancer_df['Biopsy'] #biopsy is target variable
input_df=cancer_df.drop(columns=['Biopsy'])
#target_df.shape
#input_df.shape
X=np.array(input_df).astype('float32')
Y=np.array(target_df).astype('float32')
#Y.shape
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler=StandardScaler()
X=scaler.fit_transform(X) #generating scaled version of X
X
```

```python
from sklearn.model_selection import train_test_split
#training, validation, testing data sets
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.25) #20% of dataset for testing and 80% for training
X_test, X_val, Y_test, Y_val=train_test_split(X_test,Y_test,test_size=0.5) #50% of dataset for testing and 50% for validation
```

```python
import xgboost as xgb

model=xgb.XGBClassifier(learning_rate=0.1,max_depth=50,n_estimators=100)
model.fit(X_train,Y_train)
```

```python
result_train=model.score(X_train,Y_train)
result_train #97% accuracy on training data
```

```python
#predicting score of trained model using testing dataset
result_test=model.score(X_test,Y_test)
result_test
```

```python
#making predictions on test data
Y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(Y_test,Y_predict))
```

```python
conf_mat=confusion_matrix(Y_predict,Y_test)
sns.heatmap(conf_mat,annot=True)
```
